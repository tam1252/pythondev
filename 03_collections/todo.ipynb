```python
try:
    with open("todo.txt", "r") as file:
        todo_list = file.read().splitlines()
except FileNotFoundError:
    todo_list = []

while True:
    menu = int(input("番号を入力してください[1:追加、2:表示、3:削除、4:保存と終了]"))

    match menu:
        case 1:
            task = input("TODOを入力してください: ")
            todo_list.append(task)
        case 2:
            print("TODOリスト:")
            for idx, task in enumerate(todo_list):
                print(f"{idx}. {task}")
        case 3:
            for idx, task in enumerate(todo_list):
                print(f"{idx}. {task}")
            try:
                delete_idx = int(input("削除する番号を入力してください: "))
            except ValueError:
                print("無効な入力です。")
                continue
            if 0 <= delete_idx < len(todo_list):
                todo_list.pop(delete_idx)
            else:
                print("無効な番号です。")
        case 4:
            with open("todo.txt", "w") as file:
                file.write("\n".join(todo_list))
            print("TODOリストを保存して終了します。")
            break
```
