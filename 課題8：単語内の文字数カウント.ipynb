```python
res = []
alphabet_code = {}
alphabet_code = dict.fromkeys(list("abcdefghijklmnopqrstuvwz"), 0)

while True:
    word = input("単語を入力してください（終了するには未入力でエンターを押す）: ")
    if not word:
        break
    res.append(word)
    res.sort()
    
    for w in word:
        if w in alphabet_code:
            alphabet_code[w] += 1


print(res)
for key, value in alphabet_code.items():
    if value > 0:
        print(f"{key}: {value}")
```
